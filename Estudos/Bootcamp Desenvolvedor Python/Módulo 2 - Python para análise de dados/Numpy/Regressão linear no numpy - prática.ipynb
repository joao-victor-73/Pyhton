# Regressão linear no Numpy - prática

```python
import numpy as np
import matplotlib.pyplot as plt # visualização de dados
```

```python
# Dados:
x = [-1, -0.77778, -0.555556, -0.3333333, -0.111111111, 
     0.111111111, 0.3333333, 0.555555556, 0.777777788, 1]

y = [-1.12632654, -0.57177999, -0.2169703, 0.5425699, 0.49406657, 
    1.14972239, 1.64228553, 2.1749824, 2.64773614, 2.95684201]
```

```python
# plot dos dados:
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label="Dados Originais")
plt.legend()  # Serve para mostrar a legenda acima.

plt.xlabel("X")  # Nome do eixo X
plt.ylabel("Y")  # Nome do eixo Y

plt.grid()  # é a grade do gráfico
plt.show()  # mostra o gráfico na saida do console
```

### Exemplo 1:

```python
"""
Iremos estimar uma função do tipo: y = a*x + b
ou seja, devemos achar quais os valores de a e b 
que melhor representam os dados.

os valores reais de a e b são: 2, 1
"""

# Transformando para numpy e vetor colunar:
x, y = np.array(x).reshape(-1, 1), np.array(y).reshape(-1, 1)

# Adicionando bias: para estimar o termo b
X_2 = np.hstack((x, np.ones(x.shape)))

# estimando a e b
beta = np.linalg.pinv(X_2).dot(y)

print("A estimado é: ", beta[0][0])
print("B estimado é: ", beta[1][0])
```

```python
# Plot dos dados para visualizar:
plt.figure(figsize = (10, 5))

plt.plot(x, y, 'o', label=("Dados Originais"))

plt.plot(x, X_2.dot(beta), label="Regressão Linear")
plt.legend()

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regressão linear com numpy")

plt.grid()
plt.show()
```
