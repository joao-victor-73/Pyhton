# Visualização de dados no Pandas

Além de ser escrito em cima do numpy, o Pandas também herda os métodos de visualização do _matplotlib_, uma biblioteca de visualização de dados muito versátil e utilizada. 

Alguns plots podem ser feitos com apenas uma linha de código no Pandas

```python
import pandas as pd
```

```python
# Criando arquivo/caminho do arquivo
df = pd.read_csv("https://pycourse.s3.amazonaws.com/temperature.csv")

# Definindo (setando) um novo index para o arquivo
df = df.set_index('date')

df
```


## Plot de Linhas

```python
# plot de linhas básico:
df.plot()
```

___

___

```python
# Plot de linhas com tamanho configuravel:
df.plot(figsize = (10, 5))
```

___

___

```python
# Plot de linhas com o grid (a grade para melhor visualização dos dados):
df.plot(figsize=(10, 5), grid = True)
```

___

___

```python
# Plot de linhas com style
df.plot(style='-o', figsize=(10, 5), grid = True)

# Outras formas de aparecer o ponto no gráfico:
# df.plot(style='--', figsize=(10, 5), grid = True)
# df.plot(style='-.', figsize=(10, 5), grid = True)
```

___

___

```python
# Plot de linhas com linewidth (deixa o traço mais grosso)
df.plot(style='-o', linewidth=5, figsize=(10, 5), grid = True)
```

___

___

```python
# Plot de linhas para mudar a cor da linha / traço
df.plot(style = '-o', linewidth = 2.5, color = 'red', figsize = (10, 5), grid = True)

'''
Na parte color = '' <- ele vai aceitar valores hexadecimais, como por exemplo
a cor #b05dcf
'''
```

___

___


## Plot de Barras

```python
# plot de barras básico:
df.plot(kind='bar', figsize=(10, 5), rot=0)

# rot vai servir para rotacionar os nomes do eixo X
```

___

___

- Nesse exemplo, utilizaremos a coluna de 'classification' po

- No pandas, temos um método que faz as contagens de uma determinada coluna, e também conta as Strings.
    - Esse metodo é o __value_counts()__

```python
df['classification'].value_counts()
```

```python
df['temperatura'].value_counts()
```

___

___

```python
# Fazendo um plot de barras com a contagem da coluna 'classification'
df['classification'].value_counts().plot.bar(figsize=(10, 5), 
                                          rot = 0)
```

___

___


## Gráfico de Setores (_plot de "pizza"_)

```python
# pie plot
df['classification'].value_counts().plot.pie(autopct='%1.1f%%',
                                           shadow=True,
                                           figsize=(10, 7))

# shadow = vai ser o sombreamento do gráfico;

# autopct = ele fica responsável pelas porcentagens no gráfico;
# %1.1f%% == 01.0% 
# Uma casa decimal após a vírgula/ponto e é um valor flutuante
```
